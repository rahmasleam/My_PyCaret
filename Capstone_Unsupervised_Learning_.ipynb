#K-Means Clustering

```python
X = df_pca.copy()
X
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_range = range(1, 15)
cluster_errors = []
cluster_sil_scores = []
```

```python
for num in cluster_range:
    clusters = KMeans(num, n_init = 100,random_state=0)
    clusters.fit(X)
    # capture the cluster lables
    labels = clusters.labels_
    # capture the centroids
    centroids = clusters.cluster_centers_
    # capture the intertia
    cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame({ "num_clusters":cluster_range, "cluster_errors": cluster_errors} )
clusters_df[0:10]
```

```python
plt.figure(figsize=(15,6))
plt.plot(clusters_df["num_clusters"],clusters_df["cluster_errors"],marker = 'o')
plt.xlabel('count of clusters')
plt.ylabel('error')
```

#Dimensionality Reduction with PCA and K-means Clustering

```python
# create PCA object
pca = PCA(n_components=3)
```

```python
# fit the data and transform it
features = rfm_df.copy()
pca_features = pca.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
```

```python
# create KMeans object
kmeans = KMeans(n_clusters=4)
```

```python
# fit the data and predict the clusters
clusters = kmeans.fit_predict(pca_features)
```

```python
# add the cluster labels to the original dataframe
rfm_df["Cluster"] = clusters
```

```python
products = temp.copy()
```

```python
def recommend_products(cluster_num):
    # get the cluster customers
    cluster_customers = rfm_df[rfm_df["Cluster"] == cluster_num].index.tolist()

    # get the top products bought by the cluster customers
    top_products = products[products["CustomerID"].isin(cluster_customers) == True].groupby("StockCode").agg({"Quantity": "sum"}).sort_values(by="Quantity", ascending=False).head(10)

    # # print recommendations
    print(f"Top 10 products bought by customers in cluster {cluster_num}:\n")
    # for index, row in top_products.iterrows():
    #     print(f"{index}: {row['Quantity']} purchases")

    return top_products

# example usage
recommend_products(3)
```
